```python

data = {'Easy Apply': [True, -1]}
df = pd.DataFrame(data)


print("Original DataFrame:")
print(df)

is_boolean = np.issubdtype(df['Easy Apply'].dtype, np.bool_)


print("\nDoes the 'Easy Apply' column contain boolean values? {}".format(is_boolean))


df['Easy Apply'] = df['Easy Apply'].astype(float)


print("\nDataFrame after transformation:")
print(df)
```
